1\. **Radioactive decay chain**

${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Subtract the total number of Tl atoms that decayed at each step from the Tl sample and add them to the Lead one. Plot the evolution of the two sets as a function of time  
* Repeat the exercise by means of the inverse transform method: draw 1000 random numbers from the non-uniform probability distribution $p(t)=2^{-t/\tau}\frac{\ln 2}{\tau}$ to represent the times of decay of the 1000 Tl atoms. Make a plot showing the number of atoms that have not decayed as a function of time

```python
import numpy as np
import matplotlib.pyplot as plt
#
initial = 1000
Nt      = initial
thau    = 3.052 * 60
p       = lambda t: 1 - (2 ** (-t / thau))
#
find_decayed = lambda N, t: (np.random.random(N) < p(t)).astype(int).sum()
Nl_list      = []
for _ in range(int(thau * 10)):
    Nl_list += [find_decayed(Nt, 1)]
    Nt      -= Nl_list[-1]
    #
#####
Nl_list      = np.array(Nl_list).cumsum()
Nt_list      = initial - Nl_list
half_life_x  = (Nl_list <= (initial / 2)).astype(int).sum()
half_life_y  = Nl_list[half_life_x]
#
plt.plot(Nl_list, label = "Lead")
plt.plot(Nt_list, label = "Thallium")
plt.plot(half_life_x, half_life_y, 'o', label = "half life")
plt.legend()
plt.xlabel("seconds [s]")
plt.ylabel("number of atoms")
plt.grid(True)
plt.show()
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
from   scipy import constants
import numpy as np
#
N    = int(1e7)
Z    = 79
e    = constants.e
eps0 = constants.epsilon_0
E    = e * 7.7e6
a0   = constants.physical_constants["Bohr radius"][0]
#
b        = np.sqrt((np.random.normal(0, (a0 / 100), (N, 2)) ** 2).sum(axis = 1))
theta    = 2 * np.arctan((Z * (e ** 2)) / (2 * np.pi * eps0 * E * b))
fraction = (theta > (np.pi / 2)).astype(int).sum() / N
print("The fraction of particles which bounce back is: ", int(fraction * 1e5) / 1e3, "%")
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
import numpy as np
#
N         = 1000000
V         = 2
f         = lambda x: ((np.sin(1 / (x * (2 - x)))) ** 2)
appr1     = lambda x: int(x * 100000) / 100000
appr2     = lambda x: int(x * 1000000) / 10000
points    = np.random.rand(N, 2)
f         = np.array(list(map(f, points[:, 0] * V)))
#
# HIT/MISS
integral_1     = ((points[:, 1] < f).astype(int).sum() / N) * V
p              = integral_1 / V
integral_1_err = ((p * (1 - p) * V) / N) ** 0.5
integral_1_err = appr2(integral_1_err / integral_1)
print("hit/miss method:   ", appr1(integral_1), "+-", integral_1_err, "%")
#
# MEAN VALUE
integral_2     = f.sum() * (V / N)
f_sum          = (f.sum() / N) ** 2
f2_sum         = (f ** 2).sum() / N
integral_2_err = V * (((f2_sum - f_sum) / N) ** 0.5)
integral_2_err = appr2(integral_2_err / integral_2)
print("mean value method: ", appr1(integral_2), "+-", integral_2_err, "%")
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
import numpy as np

D      = 10
N      = 10000000
points = np.random.rand(N,D)
```

```python
I = ((((points ** 2).sum(axis = 1) <= 1).astype(int).sum()) * (2 ** D) / N)
print("integral: ", I)
```

5\. **Monte Carlo integration with importance sampling** 

Calculate the value of the integral:

$$
I=\int_0^1 \frac{x^{-1/2}}{e^x+1} dx
$$

using the importance sampling method with $w(x)=1/\sqrt{x}$. You should get a result about 0.84

```python
import numpy as np
#
N  = 1000000
xs = np.random.rand(N) ** 2
I  = (2 / (np.exp(xs) + 1)).mean()
print("integral: ", I)
```
