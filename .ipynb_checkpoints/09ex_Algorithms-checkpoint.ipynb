1\. **Maximum wind speed prediction at the Sprogø station**

The exercise goal is to predict the maximum wind speed occurring every 50 years even if no measure exists for such a period. The available data are only measured over 21 years at the Sprogø meteorological station located in Denmark. 

The annual maxima are supposed to fit a normal probability density function. However such function is not going to be estimated because it gives a probability from a wind speed maxima. Finding the maximum wind speed occurring every 50 years requires the opposite approach, the result needs to be found from a defined probability. That is the quantile function role and the exercise goal will be to find it. In the current model, it is supposed that the maximum wind speed occurring every 50 years is defined as the upper 2% quantile.

By definition, the quantile function is the inverse of the cumulative distribution function. The latter describes the probability distribution of an annual maxima. In the exercise, the cumulative probability $p_i$ for a given year i is defined as $p_i = i/(N+1)$ with $N = 21$, the number of measured years. Thus it will be possible to calculate the cumulative probability of every measured wind speed maxima. From those experimental points, the scipy.interpolate module will be very useful for fitting the quantile function. Finally the 50 years maxima is going to be evaluated from the cumulative probability of the 2% quantile.

Practically, load the dataset:

```python
import numpy as np
max_speeds = np.load('max-speeds.npy')
years_nb = max_speeds.shape[0]
```

Compute then the cumulative probability $p_i$ (`cprob`) and sort the maximum speeds from the data. Use then the  UnivariateSpline from scipy.interpolate to define a quantile function and thus estimate the probabilities.

In the current model, the maximum wind speed occurring every 50 years is defined as the upper 2% quantile. As a result, the cumulative probability value will be:

```python
fifty_prob = 1. - 0.02
```

So the storm wind speed occurring every 50 years can be guessed as:

``` python
fifty_wind = quantile_func(fifty_prob)
```



````python
import numpy as np
from   scipy import stats
from   scipy import interpolate
import matplotlib.pyplot as plt
#
# load
max_speeds = np.load('max-speeds.npy')
max_speeds.sort()
years_nb   = max_speeds.shape[0]
fifty_prob = 1 - 0.02
#
# cprob
cprob = np.array([i / (years_nb + 1) for i in range(1, 22)])
qfunc = interpolate.UnivariateSpline(cprob, max_speeds, k = 3, s = 0)
maximum_wind_speed = int(qfunc(fifty_prob) * 10) / 10
print("Predicted maximum wind speed in 50 years: ", maximum_wind_speed)
````

2\. **Curve fitting of temperature in Alaska** 

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18

min: -62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58

* Plot these temperature extremes.
* Define a function that can describe min and max temperatures. 
* Fit this function to the data with scipy.optimize.curve_fit().
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

````python
import numpy as np
import matplotlib.pyplot as plt
from   scipy import optimize
#
min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
max = np.array([ 17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18])
plt.plot(min, "o", label = "min")
plt.plot(max, "o", label = "max")
plt.xlabel("Months")
plt.ylabel("T [°C]")
plt.title("temperatures as a function of time")
plt.legend()
plt.grid()
plt.show()
````

````python
a = lambda x: int(x * 1e3) / 1e3
N = 12
#
def qgauss(x, q, beta, x0, y0, y_scale):
    x  = x - x0
    def eq(x, q):
        result = ((1 + (x * (1 - q))) ** (1 / (1 - q))) if q != 1 else np.exp(x)
        return result
        #
    #####
    #
    return y_scale * (np.sqrt(beta) * eq((- beta * (x ** 2)), q)) + y0
    #
#####
xs = np.linspace(0, 11, N)
params_min, _ = optimize.curve_fit(qgauss, xs, min, p0 = [2, 0.15, 6, -70, 148])
params_max, _ = optimize.curve_fit(qgauss, xs, max, p0 = [2, 0.15, 6, 16, 60])
#
xs = np.linspace(0, 11, N * 100)
plt.plot(xs, qgauss(xs, params_min[0], params_min[1], params_min[2], params_min[3], params_min[4]), label = "fitted_min")
plt.plot(xs, qgauss(xs, params_max[0], params_max[1], params_max[2], params_max[3], params_max[4]), label = "fitted_max")
plt.plot(min, "o", label = "min")
plt.plot(max, "o", label = "max")
plt.xlabel("Months")
plt.ylabel("T [°C]")
plt.title("temperatures as a function of time")
plt.legend()
plt.grid()
plt.show()
print("fitted max:")
print("q:", a(params_max[0]), " beta:", a(params_max[1]),
      " average:", a(params_max[2]), " y0:", a(params_max[3]),
      " y_scale:", a(params_max[4]))
#
print("\nfitted min:")
print("q:", a(params_min[0]), " beta:", a(params_min[1]),
      " average:", a(params_min[2]), " y0:", a(params_min[3]),
      " y_scale:", a(params_min[4]))
#
# the fit is reasonable
````

3\. **2D minimization of a six-hump camelback function**

$$
f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2
$$

has multiple global and local minima. Find the global minima of this function.

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use numpy.meshgrid() and pylab.imshow() to find visually the regions.
* Use scipy.optimize.minimize(), optionally trying out several of its methods.

How many global minima are there, and what is the function value at those points? What happens for an initial guess of $(x, y) = (0, 0)$ ?

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#
def f(r):
    x = r[0]
    y = r[1]
    return ((4 - (2.1*(x**2)) + ((x**4)/3))*(x**2)) + (x*y) + (((4*(y**2)) - 4)*(y**2))
    #
#####
#
xpoints = np.linspace(-2, 2, 500)
ypoints = np.linspace(-1, 1, 500)
x, y    = np.meshgrid(xpoints, ypoints)
r       = [x, y]
z       = f(r)
plt.contourf(x, y, z, levels = 40)
plt.grid()
plt.title("z = f(x, y)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
#
methods = ["Nelder-Mead",
           "Powell",
           "CG",
           "BFGS",
           "Newton-CG",
           "L-BFGS-B",
           "TNC",
           "COBYLA",
           "SLSQP",
           "trust-constr",
           "dogleg",
           "trust-ncg",
           "trust-exact",
           "trust-krylov"]
#
Ps  = [(0.00, 0.00), (-1.50, 0.75), (0.00, 0.75), (1.50, 0.50), (-1.50, -0.50), (0.00, -0.75), (1.50, -0.75)]
min = []
for method in methods:
    tmp = []
    for P in Ps:
        OPT  = optimize.minimize(f, P)
        x    = OPT.x[0]
        y    = OPT.x[1]
        z    = OPT.fun
        tmp += [[x, y, z]]
        #
    #####
    #
    min += [tmp]
    #
#####
min  = np.array(min)
esit = all([(min[0] == min[i]).all() for i in range(len(min))])
#
print("If all methods give us the same result:", esit)
min  = min[0]
print("For x =", min[0][0], "and y =", min[0][1], "-> z =", min[0][2], " seems to be a saddle point (the algorithm is stucked)")
min  = min[1:,]
min  = pd.DataFrame(min, columns = ["x", "y", "z"], index = ["local", "global", "local", "local", "global", "local"])
print("\nMinima are:")
print(min)
print("\nThere are two global minima (within the range of the numerical method) in which the function (z) has the value:",
      f"{min.loc["global", "z"].iloc[0]:.6f}")
````

5\. **FFT of an image**

* Examine the provided image `moonlanding.png`, which is heavily contaminated with periodic noise. In this exercise, we aim to clean up the noise using the Fast Fourier Transform.
* Load the image using pylab.imread().
* Find and use the 2-D FFT function in scipy.fftpack, and plot the spectrum (Fourier transform of) the image. Do you have any trouble visualising the spectrum? If so, why?
* The spectrum consists of high and low frequency components. The noise is contained in the high-frequency part of the spectrum, so set some of those components to zero (use array slicing).
* Apply the inverse Fourier transform to see the resulting image.

````python
import pylab
from scipy import fftpack
import numpy as np
import matplotlib.pyplot as plt
#
# making spectrum
spectrum = lambda x, shift: np.log(np.abs(fftpack.fftshift(x)) ** 2) if shift else np.log(np.abs(x) ** 2)
#
# acting filter
def filtered(image, filter):
    filtered_image = fftpack.ifftshift(fftpack.fftshift(image) * filter)
    filtered_image[filtered_image == 0] = 1e-6
    return filtered_image
    #
#####
#
# plot
def plot_image(image, title = "spectrum"):
    plt.imshow(image)
    plt.title(title)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
    #
#####
````

````python
# reading
moonlanding = pylab.imread("moonlanding.png")
#
# plotting image
plot_image(moonlanding, "original image")
````

````python
fft2_image = fft.fft2(moonlanding)
plot_image(spectrum(fft2_image, True))
````

````python
def make_filter(D = 2, th = 0.68):
    area  = moonlanding.shape
    x_tot = area[0]
    y_tot = area[1]
    #
    def is_outside(x, y):
        if D != 1:
            return (((((x / x_tot) - 0.5) ** D) + (((y / y_tot) - 0.5) ** D)) ** (1 / D) <= (0.5 * (th / 3)))
        else:
            return (((abs((x / x_tot) - 0.5)) + (abs((y / y_tot) - 0.5))) <= (0.5 * (th / 3)))
            #
        #####
        #
    #####
    #
    xs     = np.linspace(0, x_tot - 1, x_tot)
    ys     = np.linspace(0, y_tot - 1, y_tot)
    #
    filter = np.array([[is_outside(x, y) for y in ys] for x in xs])
    return filter
    #
#####
#
def find_maxs(x):
    y              = np.abs(x)
    find_maxs.maxs = np.array(np.where(y == y.max())).T
    #
#####
#
def remove_maxs(x, frac = 0.01):
    N = int(x.shape[0] * x.shape[1] * frac)
    for _ in range(N):
        find_maxs(x)
        for max in find_maxs.maxs:
            x[max[0], max[1]] = 1e-6
            #
        #####
        #
    #####
    #
    return x
#####
````

````python
filter         = make_filter(D = 2, th = 0.65)
fft2_image_pos = filtered(fft2_image, filter)
fft2_image_neg = filtered(fft2_image, (~ filter))
fft2_image_neg = remove_maxs(fft2_image_neg.copy(), frac = 0.01)
fft2_image_new = fft2_image_pos + fft2_image_neg
````

````python
plot_image(spectrum(fft2_image_pos, True),  "pos inverted spectrum")
````

````python
plot_image(spectrum(fft2_image_pos, False), "pos original spectrum")
````

````python
plot_image(spectrum(fft2_image_neg, True),  "neg inverted spectrum")
````

````python
plot_image(spectrum(fft2_image_neg, False), "neg original spectrum")
````

````python
plot_image(spectrum(fft2_image_new, False), "tot original spectrum")
````

````python
moonlanding = fft.ifft2(fft2_image_new)
plot_image(moonlanding.real, "derived image")
````

````python
moonlanding = fft.ifft2(fft2_image_neg)
plot_image(moonlanding.real, "just noise")
````
