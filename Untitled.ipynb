```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.poisson(lam=5, size=1000)

# Calculate histogram
counts, bin_edges = np.histogram(data, bins=30)

# Calculate bin centers
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])

# Calculate Poisson uncertainty (sqrt of counts)
errors = np.sqrt(counts)

# Plotting
plt.errorbar(bin_centers, counts, yerr=errors, fmt='o', label='Data points with Poisson uncertainty', capsize=5)

# Labels and title
plt.xlabel('Value')
plt.ylabel('# of points')
plt.title('Histogram with Poisson Uncertainty')
plt.legend()

# Show the plot
plt.show()
```
